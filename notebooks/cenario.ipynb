```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)

file_path = "../data/raw/Analise_LongNeck_WSNP - Sem repostas.xlsb"

xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name="Cenário Divulgado", header=None, engine="pyxlsb")
df_data = df.iloc[:, :]
df_data
```

```python
regioes_validas = ["Mapapi", "NE Norte", "NE Sul", "NO Araguaia", "NO Centro"]

df_reg = df[df.iloc[:, 2].isin(regioes_validas)].copy()

cols_demanda = [3, 16, 27, 38]

demanda_total = pd.DataFrame()

for i, col in enumerate(cols_demanda):
    
    resumo = (
        df_reg
        .assign(valor=pd.to_numeric(df_reg.iloc[:, col], errors="coerce"))
        .groupby(df_reg.iloc[:, 2])["valor"]
        .sum()
    )
    
    demanda_total[f"W{i}"] = resumo

demanda_total.index.name = "Regiao"

demanda_total["Total_Mensal"] = demanda_total.sum(axis=1)
demanda_total.loc["Total_Semanal"] = demanda_total.sum(axis=0)

demanda_total = demanda_total.round(0).astype(int)
```

```python
regioes_validas = ["Mapapi", "NE Norte", "NE Sul", "NO Araguaia", "NO Centro"]

df_reg = df[df.iloc[:, 2].isin(regioes_validas)].copy()

cols_wsnp = [5, 18, 29, 40]

wsnp_total = pd.DataFrame()

for i, col in enumerate(cols_wsnp):
    
    resumo = (
        df_reg
        .assign(valor=pd.to_numeric(df_reg.iloc[:, col], errors="coerce"))
        .groupby(df_reg.iloc[:, 2])["valor"]
        .sum()
    )
    
    wsnp_total[f"W{i}"] = resumo

wsnp_total.index.name = "Regiao"

wsnp_total["Total_Mensal"] = wsnp_total.sum(axis=1)
wsnp_total.loc["Total_Semanal"] = wsnp_total.sum(axis=0)

wsnp_total = wsnp_total.round(0).astype(int)
```

```python
gap = wsnp_total.sub(demanda_total, fill_value=0)
gap.style.set_caption("Gap (Produção - Demanda)")
```

Análise do Gap de Produção vs Demanda – Visão Semanal

Os gráficos apresentados mostram o gap (Produção – Demanda) por região ao longo das quatro semanas do mês.  
Quando o valor está acima de zero, significa que a região possui superávit de produção e pode abastecer outras regiões. Quando está abaixo de zero, indica déficit, ou seja, necessidade de recebimento via transferência.

---

Ao longo das semanas, percebe-se um comportamento relativamente consistente:

Regiões com déficit estrutural:
- MAPAPI  
- NE Sul  
- NO Araguaia (em menor magnitude, mas recorrente)

Essas regiões apresentam insuficiência de volume para atender a nova demanda, o que indica pressão contínua ao longo do mês.

Regiões com superávit:
- NE Norte  
- NO Centro  

Essas regiões concentram a maior parte da folga produtiva e, portanto, são candidatas naturais a realizar transferências internas para equilibrar o sistema.

---

Interpretação estratégica:

O gap total mensal da regional é negativo, o que indica que a produção alocada ao Nordeste não é suficiente para absorver integralmente o aumento de demanda.
Mesmo considerando a redistribuição interna entre regiões, o sistema permanece pressionado, por isso a principal alavanca operacional disponível é a transferência inter-regional, especialmente a partir de São Paulo ou Uberlândia.

Isso implica em:

- Necessidade de um plano logístico estruturado  
- Avaliação do custo adicional de transferência  
- Análise do risco de ruptura caso os incentivos comerciais sejam mantidos  

---

Conclusão:
O aumento de demanda pressiona um sistema que já opera próximo do limite. A decisão de manter os incentivos comerciais deve considerar não apenas o ganho de market share, mas também o custo logístico adicional e o risco operacional envolvido.

```python
import pandas as pd
# ─────────────────────────────────────────────────────────────────
# 1. Localiza o bloco Malzbier pelo SKU 70792
# ─────────────────────────────────────────────────────────────────
mask = df.iloc[:, 0] == 70792.0
linha_inicio = df[mask].index[0]

# Linha de TOTAL do bloco (busca até 10 linhas abaixo)
janela = df.loc[linha_inicio : linha_inicio + 10]
linha_total = janela[janela.iloc[:, 2] == "TOTAL"].index[0]

bloco = df.loc[linha_inicio : linha_total].copy().reset_index(drop=True)

# ─────────────────────────────────────────────────────────────────
# 2. Mapeamento de colunas por semana (baseado na estrutura real)
#    Cada semana ocupa 13 colunas no layout original
# ─────────────────────────────────────────────────────────────────
SEMANAS = [
    "W0 - 02/02/2026",
    "W1 - 09/02/2026",
    "W2 - 16/02/2026",
    "W3 - 23/02/2026",
]

# Offsets das colunas dentro de cada bloco semanal
# (W0 começa na col 2; cada semana tem 13 colunas de espaço)
OFFSETS = {
    "GEO/REG":            2,
    "Demanda":            3,
    "WSNP":               5,
    "EI Semana":          7,
    "Suf.ini(d)":         8,
    "Transf. Interna":    9,
    "Transf. Ext (Cabo)": 10,
    "Transf. Ext (Rodo)": 11,
    "Trânsito":           12,
    "EF Semana":          13,
    "Suf. f (d)":         14,
}

PASSO = 13   # distância entre o início de cada semana

registros = []

for i, semana in enumerate(SEMANAS):
    deslocamento = i * PASSO

    for _, row in bloco.iterrows():
        regiao = row[OFFSETS["GEO/REG"]]
        if pd.isna(regiao) or regiao == "GEO/REG":
            continue

        reg = {"Semana": semana, "GEO/REG": regiao}
        for campo, col_base in OFFSETS.items():
            if campo == "GEO/REG":
                continue
            col_real = col_base + deslocamento
            reg[campo] = row[col_real] if col_real < len(row) else None

        registros.append(reg)

df_malzbier = pd.DataFrame(registros)

# ─────────────────────────────────────────────────────────────────
# 3. Organiza em formato pivô por semana (igual ao print)
#    Linhas = regiões | Colunas = campos da semana
# ─────────────────────────────────────────────────────────────────
CAMPOS_EXIBIR = [
    "Demanda",
    "WSNP",
    "EI Semana",
    "Suf.ini(d)",
    "Transf. Interna",
    "Transf. Ext (Cabo)",
    "Transf. Ext (Rodo)",
    "Trânsito",
    "EF Semana",
    "Suf. f (d)",
]

# Formata números
def fmt(v):
    try:
        return round(float(v), 3)
    except:
        return v

for semana in SEMANAS:
    print(f"\n{'='*80}")
    print(f"  {semana}")
    print('='*80)

    semana_df = df_malzbier[df_malzbier["Semana"] == semana][["GEO/REG"] + CAMPOS_EXIBIR].copy()
    semana_df[CAMPOS_EXIBIR] = semana_df[CAMPOS_EXIBIR].map(fmt)
    semana_df = semana_df.set_index("GEO/REG")

    # Garante que TOTAL aparece por último
    ordem = [r for r in semana_df.index if r != "TOTAL"] + ["TOTAL"]
    semana_df = semana_df.loc[ordem]

    print(semana_df.to_string())

# ─────────────────────────────────────────────────────────────────
# 4. (Opcional) Exporta para Excel com formatação
# ─────────────────────────────────────────────────────────────────
with pd.ExcelWriter("malzbier_balanca_output.xlsx", engine="openpyxl") as writer:
    for semana in SEMANAS:
        semana_df = df_malzbier[df_malzbier["Semana"] == semana][["GEO/REG"] + CAMPOS_EXIBIR].copy()
        semana_df[CAMPOS_EXIBIR] = semana_df[CAMPOS_EXIBIR].map(fmt)
        nome_aba = semana.replace("/", "-").replace(" ", "_")[:31]
        semana_df.to_excel(writer, sheet_name=nome_aba, index=False)

print("\n✅ Arquivo 'malzbier_balanca_output.xlsx' gerado com sucesso!")
```
