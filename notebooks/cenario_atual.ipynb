```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

file_path = "C:/Users/bapti/OneDrive/Documentos/Insper/Ps jr/hackathon-ambev-dados/data/raw/Analise_LongNeck_WSNP - Sem repostas.xlsb"

xls = pd.ExcelFile(file_path)
xls.sheet_names
```

```python
df = pd.read_excel(xls, sheet_name="Cenário atual BR", header=None, engine="pyxlsb")

df_data = df.iloc[:13, :]
df_data
```

```python
df_jan = df.iloc[5:13, [2,3,7,8,10,11,12,13,14]].copy()
df_jan = df_jan[df_jan[2] != "Export"]
df_jan = df_jan.reset_index(drop=True)

df_jan
```

```python
df_fev = df.iloc[5:13, [16,17,19,21,22,23]].copy()
df_fev = df_fev[df_fev[16] != "Export"]
df_fev = df_fev.reset_index(drop=True)

df_fev
```

```python
df_jan.columns = [
    "GEO/REG",
    "Demanda",
    "Prod_Real",
    "Prod_1W",
    "EI_Mes",
    "Suf_ini_d",
    "Transf_Malha",
    "EFM",
    "Suf_f_d"
]

df_fev.columns = [
    "GEO/REG",
    "Demanda",
    "WSNP",
    "Transf_Malha",
    "EFM",
    "Suf_f_d"
]
```

```python
for col in df_jan.columns[1:]:
    df_jan[col] = pd.to_numeric(df_jan[col], errors="coerce")
df_jan.dtypes

for col in df_fev.columns[1:]:
    df_fev[col] = pd.to_numeric(df_fev[col], errors="coerce")
df_fev.dtypes
```

```python
df_jan = df_jan.round()
df_jan
```

```python
df_fev = df_fev.round()
df_fev
```

```python
df_jan.columns = pd.MultiIndex.from_product(
    [["Janeiro"], df_jan.columns]
)

df_fev.columns = pd.MultiIndex.from_product(
    [["Fevereiro"], df_fev.columns]
)

df_final = pd.concat([df_jan, df_fev], axis=1)
df_final
```

```python
df_compare = pd.DataFrame({
    "Jan": df_jan[("Janeiro", "Demanda")].values,
    "Fev": df_fev[("Fevereiro", "Demanda")].values
}, index=df_jan[("Janeiro", "GEO/REG")])

df_compare["Var_abs"] = df_compare["Fev"] - df_compare["Jan"]
df_compare["Var_%"] = (df_compare["Var_abs"] / df_compare["Jan"]) * 100

df_compare.index.name = "GEO/REG"
df_compare = df_compare.reset_index()
df_compare = df_compare.round(2)

def cor_variacao(val):
    if val > 0:
        return "color: green"
    elif val < 0:
        return "color: red"
    return ""

df_compare_cor = df_compare.style \
    .format({
        "Jan": "{:,.0f}",
        "Fev": "{:,.0f}",
        "Var_abs": "{:,.0f}",
        "Var_%": "{:.2f}%"
    }) \
    .map(cor_variacao, subset=["Var_abs", "Var_%"])

df_compare_cor
```

Análise da demanda - Janeiro vs Fevereiro

Observa-se que a demanda total apresentou uma queda de 11.958 unidades no período, indicando leve desaceleração no volume consolidado entre Janeiro e Fevereiro.

A variação entretanto não foi homogênea entre as regiões. O crescimento expressivo no RJ (+19,430) e os aumentos registrados em MG e CO não foram suficientes para compensar a queda acentudada em NENO (-21.080) e a redução em SP (-8.355).

O cenário indica um mercado heterogêneo, com crescimento concentrado em algumas regiões e retração relevante em outras. A principal conclusão é que a queda consolidada foi puxada por regiões específicas (especialmente NENO), e não por um enfraquecimento uniforme da demanda.
